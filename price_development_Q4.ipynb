```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Read all the Excel files into pandas dataframes
file_dates = ['19.04.2021', '20.04.2021', '21.04.2021', '22.04.2021', '23.04.2021', '24.04.2021', '25.04.2021']
dfs = []

for date in file_dates:
    file_path = f'VP_{date}.xlsx'
    df = pd.read_excel(file_path)
    dfs.append(df)
```

```python
# Step 2: Concatenate the dataframes into a single dataframe
combined_df = pd.concat(dfs)
```

```python
# Step 3: Calculate the mean of the prices
price_trends = combined_df.groupby(['Datum', 'Shop ID'])['Verkaufspreis'].mean().unstack()
```

```python
print(price_trends)
```

```python
# Step 4: Visualization
price_trends.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Price Development 19.04-25.04.2021')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
